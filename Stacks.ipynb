```python
class Node:
    def __init__(self, value):
        # Initialize a node with a given value
        self.value = value
        # Pointer to the next node, initially set to None
        self.next = None
        

class Stack:
    def __init__(self, value):
        # Create a new node with the given value
        new_node = Node(value)
        # Set the top of the stack to the new node
        self.top = new_node
        # Initialize the height of the stack to 1
        self.height = 1

    def print_stack(self):
        # Print all the values in the stack
        temp = self.top
        while temp is not None:
            print(temp.value)
            temp = temp.next

    def push(self, value):
        # Create a new node with the given value
        new_node = Node(value)
        if self.height == 0:
            # If the stack is empty, set the top to the new node
            self.top = new_node
        else:
            # Push the new node onto the stack and update the top
            new_node.next = self.top
            self.top = new_node
        # Increment the height of the stack
        self.height += 1
        return True
    
    
    def pop(self):
        # Check if the stack is empty
        if self.top == None:
            return None
        else:
            # Save the top node
            temp = self.top
            # Move the top pointer to the next node
            self.top = self.top.next
            # Disconnect the saved node from the stack
            temp.next = None
            # Decrement the height of the stack
            self.height -= 1
            # Return the popped node
            return temp

# Test the Stack implementation
my_stack = Stack(4)
my_stack.push(3)
my_stack.push(2)
my_stack.push(1)

print('Stack before pop():')
my_stack.print_stack()

print('\nPopped node:')
print(my_stack.pop().value)

print('\nStack after pop():')
my_stack.print_stack()
```
